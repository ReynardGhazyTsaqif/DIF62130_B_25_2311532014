```python
import pandas as pd

data = {'luas' : [6, 8, 10, 14, 18],
      'harga' : [7, 9, 13, 17.5, 18]}

df = pd.DataFrame(data)
df
```

1. Import pustaka pandas
2. Membuat dictionary data

  Kode ini membuat dictionary bernama data dengan dua key:
  
  'luas' berisi daftar ukuran luas dalam satuan tertentu: [6, 8, 10, 14, 18]

  'harga' berisi daftar harga yang sesuai dengan luas: [7, 9, 13, 17.5, 18]
3. Menampilkan DataFrame

```python
import matplotlib.pyplot as plt

df.plot(kind='scatter', x='luas', y='harga')
plt.title('Perbandingan luas tanah dengan harga tanah')
plt.xlabel('Luas Tanah (m²)')
plt.ylabel('Harga Tanah (Dollar)')
plt.xlim(0, 25)
plt.ylim(0, 25)
plt.grid(True)
plt.show()
```

1. Import pustaka Matplotlib untuk membuat grafik.
2. Membuat Scatter Plot dengan:

  'luas' sebagai sumbu x (Luas Tanah dalam m²)

  'harga' sebagai sumbu y (Harga Tanah dalam Dollar)
3. Menambahkan Judul dan Label Sumbu

  Judul: Perbandingan luas tanah dengan harga tanah

  Label sumbu X: Luas Tanah (m²)

  Label sumbu Y: Harga Tanah (Dollar)
4. Mengatur Rentang Sumbu X dan Y dari 0 hingga 25.
5. Menambahkan Grid untuk mempermudah membaca grafik

```python
import numpy as np

x=np.array(df['luas'])
y=np.array(df['harga'])

print(f'X: {x}')
print(f'Y: {y}')
```

1. Import pustaka NumPy untuk komputasi numerik
2. Konversi Kolom 'luas' dan 'harga' Menjadi Array NumPy

  df['luas']: Mengambil kolom 'luas' dari DataFrame df (berisi nilai [6, 8, 10, 14, 18]).

  df['harga']: Mengambil kolom 'harga' dari DataFrame df (berisi nilai [7, 9, 13, 17.5, 18]).

  np.array(...): Mengonversi masing-masing kolom menjadi array NumPy.
3. Menampilkan hasil array

```python
x = x.reshape(-1, 1)

print(f'X shape: {x.shape}')
print(f'X: {x}')
```

Mengubah bentuk array x menjadi array 2D (matriks kolom).
Sebelum reshape

x = np.array([6, 8, 10, 14, 18])

Setelah reshape (2D array)

x = np.array
([[ 6],

              [ 8],
              [10],
              [14],
              [18]])
              
  Bentuk (5, 1)


```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x, y)
```

1. Impor LinearRegression dari pustaka scikit-learn, untuk membuat model regresi linier.
2. Membuat Model Regresi Linier
3. Melatih Model dengan Data

  X: Data luas tanaha
  
  Y : Data harga tanah

```python
x_vis = np.array([0,25]).reshape(-1, 1)
y_vis = model.predict(x_vis)

plt.scatter(x, y)
plt.plot(x_vis, y_vis, color='red')
plt.title('Perbandingan luas tanah dengan harga tanah')
plt.xlabel('Luas Tanah (m²)')
plt.ylabel('Harga Tanah (Dollar)')
plt.grid(True)
plt.show()
```

1. Menentukan Rentang Luas Tanah

  Data x_vis dibuat dengan nilai 0 dan 25 untuk menunjukkan batas awal dan akhir grafik.

  Model regresi digunakan untuk menghitung prediksi harga (y_vis) berdasarkan x_vis.

2. Menampilkan Data Asli

  Scatter plot untuk menampilkan data asli dalam bentuk titik-titik.

3. Menampilkan Garis Regresi Linier

4. Menambahkan Informasi Tambahan

  Judul, label sumbu, dan grid ditambahkan untuk memperjelas grafik

```python
print(f'Model Intercept: {model.intercept_}')
print(f'Slope: {model.coef_}')
```

1. Intercept (model.intercept_)

  nilai harga tanah saat luas tanah = 0.
titik awal garis pada sumbu y.

2. Slope (model.coef_)

  kemiringan garis regresi, yang menunjukkan seberapa besar harga tanah berubah ketika luas tanah bertambah 1 unit.

```python
luas_tanah = np.array([12,15,20]).reshape(-1, 1)
print(luas_tanah.shape)
luas_tanah
```

Mmembuat array 2D yang berisi nilai luas tanah 12, 15, dan 20.
Kemudian array ditampilkan dengan format tegak lurus

```python
harga_prediksi = model.predict(luas_tanah)
harga_prediksi
```

Menggunakan model regresi linier untuk memprediksi harga tanah berdasarkan nilai luas tanah (12, 15, dan 20).

```python
for luas, hrg in zip(luas_tanah, harga_prediksi):
  print(f'Luas Tanah {luas}: harga prediksi: {hrg}')
```

melakukan iterasi melalui pasangan luas tanah dan harga prediksi menggunakan zip().

Pada setiap iterasi, kode mencetak luas tanah beserta harga tanah yang diprediksi oleh model regresi linier. Sehingga tampilannya akan lebih mudah dilihat

```python
x_train = np.array([6, 8, 10, 14, 18]).reshape(-1, 1)
y_train = np.array([7, 9, 13, 17.5, 18])

x_test = np.array([9, 9, 11, 16, 12]).reshape(-1, 1)
y_test = np.array([11, 8.5, 15, 18, 11])
```

membuat data pelatihan (x_train, y_train) dan data pengujian (x_test, y_test) untuk model regresi linier. Tiap data pelatihan dan data pengujian berisi nilai-nilai tersendiri

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

membuat model regresi linier dan melatihnya menggunakan data pelatihan (x_train, y_train).

```python
from sklearn.metrics import r2_score

y_pred = model.predict(x_test)
r_squared = r2_score(y_test, y_pred)
print(f'R-squared: {r_squared}')
```

menghitung koefisien determinasi (R² score) untuk mengevaluasi seberapa baik model regresi linier memprediksi data uji.

1. Membuat Prediksi

  model.predict(x_test) menghasilkan prediksi harga tanah (y_pred) berdasarkan luas tanah dalam x_test.

2. Menghitung R² Score
  
  r2_score(y_test, y_pred) membandingkan nilai aktual (y_test) dengan nilai prediksi (y_pred).
